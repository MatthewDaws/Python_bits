# Sampling from a "complex" probability distribution

A (pseudo)-random number generator can easily give you samples from a uniform distribution.  How do we get samples from e.g. a Gaussian?

Let's think generally for a moment.  If $Y \sim U(0,1)$ and $(Y_i)_{i\geq 1}$ are iid copies of $Y$, then given samples $(y_i)$ we expect, for any $0 < p < 1$, that the proportion $\frac1n |\{ i \leq n : y_i \leq p \}| \rightarrow p$ as $n\rightarrow\infty$.  If $X$ has [CDF](https://en.wikipedia.org/wiki/Cumulative_distribution_function) $F_X$ then set
$$ Z = \inf \{ x : F_X(x) \geq Y \} = \min \{ x : F_X(x) \geq Y \} $$
as $F_X$ is right-continuous and non-decreasing.  Then
$$ \mathbb P(Z \leq x) = \mathbb P( \min \{ x' : F_X(x') \geq Y \} \leq x )
= \mathbb P( F_X(x) \geq Y ) = F_X(x) $$
as $Y$ is uniform on $(0,1)$, and again using that $F_X$ is right-continuous and non-decreasing.

So, $F_X$ is continuous and increasing between 0 and 1, then $F_X^{-1} : (0,1) \rightarrow \mathbb R$ exists and $Z = F_X^{-1}(Y)$ (we don't particularly worry about what happens if $Y=0$ or $1$ as these are null sets).

To pick a nice example which works, consider the [Exponential distribution](https://en.wikipedia.org/wiki/Exponential_distribution) which has $F_X(x) = 1 - \exp(-\lambda x)$ where $\lambda > 0$ is the "rate".  So given a sample $y$ from $U(0,1)$ we set $x = F_X^{-1}(y) = -\frac{\log(1-y)}{\lambda}$ or by symmetry, $x = \frac{-1}{\lambda}\log(y)$.

```python
num_samples = 10000
y = np.random.uniform(size=num_samples)
samples = -np.log(y) / 3
fig, axes = plt.subplots(figsize=(10,4))
axes.hist(samples, bins=np.arange(0, 3.1, 0.2))
x = np.arange(0, 3.1, 0.2)
y = 3 * np.exp(-3*x)
axes.plot(x, y*num_samples/6)
```

## Rejection methods

Here, suppose we have random variables $X$ and $Y$, where $X$ has a continuous probability density function (PDF) $f$ which is complicated to sample from, but $Y$ has a CDF $g$ from which we can sample from.  (We are allowed that $f$ and $g$ are not normalised, so $\int_{-\infty}^\infty f = Z_f$ is by definition positive, but may not be 1.)  We assume further that we know some constant $c$ with $f(x) < cg(x)$ for all $x$.

   - Then let $y$ be a sample from $Y$ and $u$ be a sample from $U(0,1)$.  Set $v = u c f(y)$.
   - If $f(y) \geq v$ then $y$ is _accepted_ as a sample $x=y$ of $X$
   - Otherwise we _reject_ and go back to the start.
   
For example, let $f(x) = \exp(-x^2/10) + \exp(-(x-2)^2/2)$ an un-normalised mixture of Gaussians.  Let's use $Y$ as a Gaussian (yes, yes, we haven't fully explored how to sample from $Y$ yet...) with mean 2 and $\sigma^2 = 10$, so $g(x) = \exp(-(x-2)^2 / 20$.  From trial and error we find that $c=2$ works:

```python
def f(x):
    return np.exp(-x**2 / 10) + np.exp(-(x-2)**2 / 2)

def g(x):
    return np.exp(-(x-2)**2 / 20)

c = 2.0
x = np.arange(-6, 10.1, 0.1)
fig, axes = plt.subplots(figsize=(10,4))
axes.fill_between(x, 0, g(x) * c, color="green", alpha = 0.4)
axes.plot(x, g(x) * c, lw=2, color="red")
axes.fill_between(x, 0, f(x), color="green", alpha = 0.8)
axes.plot(x, f(x), lw=2, color="red")
```

The method works because of random point $(y,v) \in\mathbb R^2$ is randomly chosen from then light green shaded area.  We accept the sample precisesly when this point lands in the dark green area, and reject otherwise.  Thus the accepted points are uniformly chosen from the dark green area, and so the generated $x$ are chosen with probability proportional to $f(x)$, i.e. from the normalised distribution given by $f$.

To test this, we note that $f(x) = \frac{\sqrt{10\pi}}{\sqrt{10\pi}} \exp(-x^2/10) + \frac{\sqrt{2\pi}}{\sqrt{2\pi}} \exp(-(x-2)^2/2)$ so this is a mixture of $\sqrt{10\pi}$ lots of $N(0,5)$ and $\sqrt{2\pi}$ lots of $N(2,1)$.

```python
def gen_from_f():
    while True:
        y = np.random.normal(2, np.sqrt(10))
        u = np.random.uniform(0,1)
        v = u * g(y) * 2
        if f(y) >= v:
            return y
        
def true_dist():
    r1 = np.sqrt(10*np.pi)
    r2 = np.sqrt(2*np.pi)
    if np.random.uniform()*(r1+r2) <= r1:
        return np.random.normal(0,np.sqrt(5))
    else:
        return np.random.normal(2,1)

samples = [gen_from_f() for _ in range(20000)]
data = [true_dist() for _ in range(20000)]
fig, axes = plt.subplots(figsize=(10,4))
bins = np.arange(-6,5.001,0.2)
axes.set_xlim([-6, 5])
_ = axes.hist(data, bins=bins, cumulative=True, alpha=0.5)
_ = axes.hist(samples, bins=bins, cumulative=True, alpha=0.5)
```

We see the there is very little difference in the cumulative histograms, exactly as we hope!

# Metropolis-Hastings

This is a [MCMC Method](https://en.wikipedia.org/wiki/Markov_chain_Monte_Carlo) or Random Walk method.  We execute a random walk (technically, a Markov Chain) with the long-term aim of jumping to a point $x$ with probability about $f(x) / Z$ (where $Z$ is the normalisation of $f$, which again we typically don't know).

Let me stick to the Metropolis case first.  In our random walk, let $Q(x|y)$ be the probability of jumping from $y$ to $x$, and we'll assume $Q(x|y) = Q(y|x)$.  E.g. $Q(\cdot|y) \sim N(y,\sigma^2)$.

   - Suppose we're currently at $y$.
   - Sample $x'$ from $Q(\cdot|y)$
   - Compute $\alpha = f(x') / f(y)$.  (Notice that $Z$ would cancel out here).
   - If $\alpha \geq 1$ then $x'$ is more probable than $y$, so jump to $x'$.
   - Otherwise, with probability $\alpha$ just to $x'$ anyway.  Otherwise stay at $y$.
   
The well-known problems with the Metropolis-Hastings method is that (a) the distribution of $x$ is only _aymptotically_ of distribution $f$, so a "burnt-in" period is nedeed; (b) the choice of $Q$ can be tricky: if the jump size is wrong, then you can get very slow exploration of the sample space, and large correlations between successive samples (meaning you end up looking at e.g. only 1 in 100 samples generated).

   - You can see this in the following code: In the `metropolis` function, change the variance of the normal distribution: a value of `0.1` leads to signifcantly worse results (effectively wrong results, in fact).

See [Wikipedia](https://en.wikipedia.org/wiki/Metropolis%E2%80%93Hastings_algorithm) for further details.

```python
def f(x):
    return np.exp(-(x+2)**2 / 5) + np.exp(-(x-2)**2 / 2)

def true_dist():
    r1 = np.sqrt(5*np.pi)
    r2 = np.sqrt(2*np.pi)
    if np.random.uniform()*(r1+r2) <= r1:
        return np.random.normal(-2,np.sqrt(2.5))
    else:
        return np.random.normal(2,1)

x = np.arange(-6, 6.1, 0.1)
fig, axes = plt.subplots(figsize=(10,4))
axes.fill_between(x, 0, f(x), color="green", alpha = 0.8)
_ = axes.plot(x, f(x), lw=2, color="red")
```

```python
# Should return a sample from $Q(\cdot|y)$
def Q(y):
    return y + np.random.normal(0,2)

def metropolis(y, step = 1):
    for _ in range(step):
        x = Q(y)
        probnew, probold = f(x), f(y)
        if probnew >= probold or np.random.uniform() * probold <= probnew:
            y = x
    return y

x = 0
path = []
for _ in range(1000):
    x = metropolis(x)
    path.append(x)
    
fig, axes = plt.subplots(figsize=(15,4))
axes.plot(path)
```

```python
# Burnt-in
x = metropolis(0, 10000)
# Generate samples in a naive way
samples = []
for _ in range(10000):
    x = metropolis(x, step=1)
    samples.append(x)
    
data = [true_dist() for _ in range(10000)]
fig, axes = plt.subplots(figsize=(10,4))
bins = np.arange(-6,5.001,0.2)
axes.set_xlim([-6, 5])
_ = axes.hist(data, bins=bins, cumulative=True, color = "green", alpha=0.5)
_ = axes.hist(samples, bins=bins, cumulative=True, color = "blue", alpha=0.5)
```

```python
# Burnt-in
x = metropolis(0, 10000)
# Generate samples by taking only one every 10.
samples = []
for _ in range(10000):
    x = metropolis(x, step=10)
    samples.append(x)
    
data = [true_dist() for _ in range(10000)]
fig, axes = plt.subplots(figsize=(10,4))
bins = np.arange(-6,5.001,0.2)
axes.set_xlim([-6, 5])
_ = axes.hist(data, bins=bins, cumulative=True, color = "green", alpha=0.5)
_ = axes.hist(samples, bins=bins, cumulative=True, color = "blue", alpha=0.5)
```
